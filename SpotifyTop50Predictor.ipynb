```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/top50.csv', encoding='latin-1')
```

```python
df
```

```python
df.info()
```

## Preprocessing

```python
df = df.drop(['Unnamed: 0', 'Track.Name'], axis=1)
```

```python
df
```

```python
df['Popularity'] = pd.qcut(df['Popularity'], q=2, labels=[0, 1])
```

```python
df['Popularity']
```

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
df = onehot_encode(df, 'Genre', 'genre')
df = onehot_encode(df, 'Artist.Name', 'artist')
```

```python
y = df.loc[:, 'Popularity']
X = df.drop('Popularity', axis=1)
scaler = StandardScaler()

X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.85, random_state=20)
```

Modelling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```

```python
log_model = LogisticRegression()
knn_model = KNeighborsClassifier()
dec_model = DecisionTreeClassifier()
mlp_model = MLPClassifier()
svm_model = SVC()
log_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
dec_model.fit(X_train, y_train)
mlp_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
log_acc = log_model.score(X_test, y_test)
knn_acc = knn_model.score(X_test, y_test)
dec_acc = dec_model.score(X_test, y_test)
mlp_acc = mlp_model.score(X_test, y_test)
svm_acc = svm_model.score(X_test, y_test)
```

```python
print("Logistic Regression Acc.:", log_acc)
print("K-Nearest-Neighbors Acc.:", knn_acc)
print("Decision Tree Acc.:", dec_acc)
print("Neural Network Acc.:", mlp_acc)
print("Support Vector Machine Acc.:", svm_acc)
```
